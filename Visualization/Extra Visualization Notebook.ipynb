```python
import plotly.graph_objects as go

# Function to create the first barplot
def create_barplot_1(ref_data, level_value):
    filtered_data = [d for d in ref_data if d['level'] == level_value or d['type'] == f'Total {level_value}']
    x_values = [d['desi'] for d in filtered_data]
    y_values_total = [d['total'] for d in filtered_data]
    y_values_alerts = [d['n_alerts'] for d in filtered_data]

    # Defining a custom color scale for the bars
    colors = ['rgb(31, 119, 180)', 'rgb(255, 127, 14)', 'rgb(44, 160, 44)', 'rgb(214, 39, 40)', 'rgb(148, 103, 189)',
              'rgb(140, 86, 75)', 'rgb(227, 119, 194)', 'rgb(127, 127, 127)', 'rgb(188, 189, 34)', 'rgb(23, 190, 207)']

    fig = go.Figure()
    for i, (x, y_total, y_alerts) in enumerate(zip(x_values, y_values_total, y_values_alerts)):
        fig.add_trace(go.Bar(x=[x], y=[y_total], text=[f'{y_total:.2f}'], name='Total', marker=dict(color=colors[i % len(colors)]), textposition='auto'))
        fig.add_trace(go.Bar(x=[x], y=[y_alerts], text=[f'{y_alerts}'], name='Alerts', marker=dict(color=colors[(i + 1) % len(colors)]), textposition='auto'))

    fig.update_layout(title=f'Totals and Alerts for {level_value}',
                      xaxis_title='Designation',
                      yaxis_title='Value',
                      plot_bgcolor='rgba(0,0,0,0)',
                      paper_bgcolor='rgba(0,0,0,0)',
                      font=dict(color='red', size=14),
                      legend=dict(font=dict(size=14, color='black')),
                      xaxis=dict(tickfont=dict(size=14)),
                      yaxis=dict(tickfont=dict(size=14)),
                      template="plotly_dark")

    return fig

# Function to create the second barplot
def create_barplot_2(summ_data):
    names = [d['name'] for d in summ_data]
    ref_total = [d['ref_total'] for d in summ_data]
    bids_totals = [d['bids_totals'] for d in summ_data]

    # Defining a custom color scale for the bars
    colors = ['rgb(31, 119, 180)', 'rgb(255, 127, 14)', 'rgb(44, 160, 44)', 'rgb(214, 39, 40)', 'rgb(148, 103, 189)',
              'rgb(140, 86, 75)', 'rgb(227, 119, 194)', 'rgb(127, 127, 127)', 'rgb(188, 189, 34)', 'rgb(23, 190, 207)']

    fig = go.Figure()
    for i, (name, total) in enumerate(zip(names, bids_totals)):
        fig.add_trace(go.Bar(x=[name], y=[ref_total[i]], text=[f'{ref_total[i]:.2f}'], name='Reference Total', marker=dict(color=colors[i % len(colors)]), textposition='auto'))
        for j, t in enumerate(total):
            fig.add_trace(go.Bar(x=[name], y=[t], text=[f'{t:.2f}'], name=f'Total {j+1}', marker=dict(color=colors[(i + j + 1) % len(colors)]), textposition='auto'))

    fig.update_layout(title='Reference Total and Bid Totals',
                      xaxis_title='Names',
                      yaxis_title='Value',
                      plot_bgcolor='rgba(0,0,0,0)',
                      paper_bgcolor='rgba(0,0,0,0)',
                      font=dict(color='red', size=14),
                      legend=dict(font=dict(size=14, color='black')),
                      xaxis=dict(tickfont=dict(size=14)),
                      yaxis=dict(tickfont=dict(size=14)),
                      template="plotly_dark")

    return fig

# Example data for ref_data and summ_data
ref_data = [
    {"desi": "Designation 1", "level": "Section", "type": "Total section", "total": 100.0, "n_alerts": 5},
    {"desi": "Designation 2", "level": "Groupe", "type": "Total groupe", "total": 200.0, "n_alerts": 10},
    {"desi": "Designation 3", "level": "Poste", "type": "Total poste", "total": 150.0, "n_alerts": 8},
    {"desi": "Designation 4", "level": "Sous-poste", "type": "Total sous-poste", "total": 120.0, "n_alerts": 6}
]

summ_data = [
    {"name": "Name 1", "n_bids": 5, "ref_total": 500.0, "bids_totals": [100.0, 150.0, 200.0, 120.0, 130.0]},
    {"name": "Name 2", "n_bids": 3, "ref_total": 350.0, "bids_totals": [80.0, 120.0, 150.0]},
    {"name": "Name 3", "n_bids": 7, "ref_total": 700.0, "bids_totals": [120.0, 180.0, 220.0, 80.0, 70.0, 20.0, 10.0]}
]

# Example usage to create the plots
level_value = "Section"  # You can change this value as needed
fig1 = create_barplot_1(ref_data, level_value)
fig2 = create_barplot_2(summ_data)

# Show the plots
fig1.show()
fig2.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define class labels
classes = ['Positive', 'Negative']

# Define confusion matrix labels as strings
confusion_matrix_str = [['TP', 'FN'], ['FP', 'TN']]


# Create a colorful heatmap with a custom dark color, increased font size, bold text, and bold titles
plt.figure(figsize=(6, 4))
sns.set(font_scale=2.2)  # Adjust font size)  # Set font weight to bold for all titles
heatmap = sns.heatmap([[0, 0], [0, 0]], annot=confusion_matrix_str, fmt='', cmap='YlGnBu', xticklabels=classes, yticklabels=classes, cbar=False, annot_kws={"weight": "bold"})
heatmap.set_xlabel('Predicted Label', weight='bold')  # Set x-axis label to bold
heatmap.set_ylabel('True Label', weight='bold')  # Set y-axis label to bold
plt.title('Confusion Matrix', weight='bold')  # Set title to bold
plt.savefig("cm.png",dpi=400)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sample data (replace with your own data)
tasks = ['Project Planning', 'Data Collection', 'Data Preprocessing', 
         'Feature Engineering', 'Embedded Techniques', 'FRNN Classification',
         'Web Integration', 'Documentation']
durations = [60, 20, 20, 10, 80, 30, 75, 50]  # Duration of each task in days

# Calculate average duration
avg_duration = np.mean(durations)

# Calculate total duration
total_duration = sum(durations)

# Create DataFrame
data = pd.DataFrame({'Tasks': tasks, 'Duration (Days)': durations})

# Set style
sns.set_style("whitegrid")

# Create bar plot
plt.figure(figsize=(17, 8))  # Adjust size
sns.barplot(x='Duration (Days)', y='Tasks', data=data, palette='hsv')

# Add labels and title with custom fonts
plt.xlabel('Duration (Days)', fontsize=14, fontweight='bold')
plt.ylabel('Tasks', fontsize=14, fontweight='bold')
plt.title('FYP Project Timeline', fontsize=16, fontweight='bold')

# Add value labels to bars
for index, row in data.iterrows():
    plt.text(row['Duration (Days)'] + 1, index, str(row['Duration (Days)']) + ' days', va='center', fontsize=12)

# Add legend representing duration
colors = sns.color_palette('Blues_d', len(tasks))
legend_labels = [f"{task}: {duration} days" for task, duration in zip(tasks, durations)]
plt.legend(labels=legend_labels, loc='center right')

# Add a grid for better readability
plt.grid(True, linestyle='--', alpha=0.7)

# Add average duration line
plt.axvline(x=avg_duration, color='red', linestyle='--', label=f'Average Duration: {avg_duration:.2f} days')
plt.legend(loc='upper right')


# Add a note about the duration variability
plt.text(0, -1, 'Note: Duration may vary', fontsize=18, color='gray', ha='left')

# Show plot
plt.savefig("FYP-Project-Timeline")
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

def label_encode_categorical(train_df, test_df):
    """
    Apply label encoding to categorical features in both train and test datasets.
    
    Parameters:
    train_df (pd.DataFrame): Training dataset
    test_df (pd.DataFrame): Test dataset
    
    Returns:
    pd.DataFrame: Encoded training dataset
    pd.DataFrame: Encoded test dataset
    dict: Dictionary containing the mappings of original categorical values to encoded values
    """
    # Identify categorical columns in both datasets
    cat_columns_train = train_df.select_dtypes(exclude=np.number).columns
    cat_columns_test = test_df.select_dtypes(exclude=np.number).columns

    # Initialize the LabelEncoder
    label_encoder = LabelEncoder()

    # Dictionary to store the mappings
    mappings = {}

    # Apply label encoding to categorical columns in both datasets
    for column in cat_columns_train:
        # Fit the label encoder on the combined data
        combined_data = pd.concat([train_df[column], test_df[column]], axis=0)
        label_encoder.fit(combined_data)

        # Store the mapping
        mappings[column] = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))

        # Transform the data in train dataset
        train_df[column] = label_encoder.transform(train_df[column])

        # Transform the data in test dataset
        test_df[column] = label_encoder.transform(test_df[column])

    return train_df, test_df, mappings

# Example usage:
# Assuming `train_df` and `test_df` are already defined
encoded_train, encoded_test, value_mappings = label_encode_categorical(train_dataset_ind, test_dataset_ind)

# Print the value mappings
for column, mapping in value_mappings.items():
    print(f"Column: {column}")
    for original_value, encoded_value in mapping.items():
        print(f"    {original_value} -> {encoded_value}")
```
